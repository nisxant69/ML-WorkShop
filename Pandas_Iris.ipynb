# Handling the Missing Values

```python
student = ["student1", "student2", "student1", "student4", "student5", "student1", "student7"]
age = [21, 21, 21, 19, np.nan, 21, np.nan]
gender = ["male", "Female", "male", "Other", "TungTung", np.nan, "male"]
marks = [45, 54, 45, 64, np.nan, 45, 56]
df_student = pd.DataFrame({"Students": student, "Age": age, "Gender": gender, "Marks_Obtained": marks})
df_student
```

```python
df_student.duplicated()
```

```python
df_student.duplicated().sum()
```

```python
# df_student.drop_duplicates() #ya inplace true garena vane feri same data aauxa so tesailey inplace=true garo vane feri df_student garda update hunxa
df_student.drop_duplicates(inplace = True)
```

```python
df_student
```

```python
df_student.isnull().sum()
```

# Techniques to Handle Missing Value

### First Method: Remove the row that contains missing value

```python
#dropna()
df_student.dropna()
# this remove the whole row if any column contains NAN
```

### Second Method: Forward Filling and Backward Filling

```python
df_student
```

```python
#fillna()
#forwardfill ho yo yesma chai NaN ma value fill mathi ko value bata aauxa
df_student.fillna(method='ffill')
```

```python
#fillna()
#backward ho yo yesma chai NaN ma value fill tala ko value bata aauxa
df_student.fillna(method='bfill')
```

```python
df_student.describe()
```

```python
df_student['Gender'].value_counts()
```

### Method Three: Mean, Median and Mode Imputation

```python
mean_age = df_student["Age"].mean()
mean_age
```

```python
mean_marks = df_student["Marks_Obtained"].mean()
mean_marks
```

```python
# gender_mode = df_student["Gender"].mode() #gives the answer in series way to overcome it......>
gender_mode = df_student["Gender"].mode()[0]
gender_mode
```

```python
df_filled = df_student.copy()
df_filled
```

```python
df_filled["Age"] = df_filled["Age"].fillna(mean_age)
df_filled
```

```python
df_filled["Marks_Obtained"] = df_filled["Marks_Obtained"].fillna(mean_marks)
df_filled
```

```python
df_filled["Gender"] = df_filled["Gender"].fillna(gender_mode)
df_filled
```
