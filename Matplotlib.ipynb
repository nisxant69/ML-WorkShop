# Scatter Plot

```python
# x = np.arange(1, 21)
#y = (x**3.2) + 5 #parabola
y = (x*3.2) + 5 #
# y= m*x + c , where m= slope and  C are weights and bias respectively
plt.scatter(x,y,color='g')
plt.plot(x,y)
plt.show()
```

```python
# Scatter plot using random
# x = number from 1-20
# y = Random number from 1-20 
x = np.arange(1,21)
y = np.random.randint(1,20,20)
plt.scatter(x,y,color='r')
plt.plot(x,y,color='g')
plt.show()
```

```python
plt.Figure(figsize = (1,5))
plt.scatter(x, y, color = 'r')
plt.xlabel("X-Value")            # X AXIS KO NAM RAKHNA

plt.ylabel("Y-Value")            # Y AXIS KO NAM RAKHNA
plt.title("Title: X vs Y")       # Title Set garne
plt.savefig('scatterplot')       #Locally save garna lai
plt.show()
```

# Load Iris Dataset

```python
iris=pd.read_csv('Iris.csv')
iris.head()
```

```python
iris['Species'].unique()
```

```python
iris['Species'].value_counts()
```

```python
plt.bar(iris["Species"].unique(), iris["Species"].value_counts(), color = ['r', 'b', 'g'])
plt.xlabel("Species")
plt.ylabel("Counts")
# plt.grid()
plt.show()
```

# Boxplot

```python
plt.boxplot(iris["SepalWidthCm"], notch = True, patch_artist = True, labels = ['SepalWidthCm'])
plt.show()
```

# Using seaborn

```python
sns.boxplot( data = iris, x = 'Species', y = 'SepalWidthCm', hue = 'Species')
plt.show()
```

```python
sns.scatterplot(data = iris, y='SepalLengthCm', x='Species', hue='Species')
plt.show()
```

```python
sns.scatterplot(data = iris, y='PetalLengthCm', x='Species', hue='Species')
plt.show()
```

```python
sns.barplot(data = iris, y='PetalLengthCm', x='Species', hue='Species')
plt.show()
```
